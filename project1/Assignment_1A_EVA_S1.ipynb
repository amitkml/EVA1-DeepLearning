````
# import the necessary packages
import numpy as np
from urllib.request import urlopen
import cv2
from matplotlib import pyplot as plt

# METHOD #1: OpenCV, NumPy, and urllib
def url_to_image(url):
	# download the image, convert it to a NumPy array, and then read
	# it into OpenCV format
	resp = urlopen(url)
	image = np.asarray(bytearray(resp.read()), dtype="uint8")
	image = cv2.imdecode(image, cv2.IMREAD_COLOR)
 
	# return the image
	return image
````

````
image = url_to_image("https://fontsarena-cd5e.kxcdn.com/wp-content/uploads/2019/04/helvetica-now-font-400x364.png")
````

##**Vertical Edge Detection**

**Kernel:  **

```
            [1 ,0 ,-1
            -1 ,0 ,1    
            -1, 0, 1]
```

**Process :**

You can see in the kernel that vertical edge values are positive .

````
# Vertical edge detection 

kernel = np.float32([[1,0,-1],[-1,0,1],[-1,0,1]])

dst = cv2.filter2D(image,-1,kernel)
cv2_imshow(dst)
````

##**Horizontal Edge Detector**   :  


**Kernel:  **

```
            [-1,-2,-1
              0 ,0 ,0    
              1, 2, 1]
```

**Process :**

You can see in the kernel that horizontal values are positive .


````
# sobel horizontal edge detector
kernel = np.float32([[-1,-2,-1],[0,0,0],[1,2,1]])

dst= cv2.filter2D(image,-1,kernel)
cv2_imshow(dst)
````

##**45 Degree Angle Detector **

**Kernel:  **

```
            [-1,-1,2
             -1,2,-1    
             2,-1,-1]
```

**Process :**

You can see in the kernel that diagonal values are positive .

````
kernel = np.float32([[-1,-1,2],
                    [-1,2,-1],
                    [2,-1,-1]])

dst= cv2.filter2D(image,-1,kernel)
cv2_imshow(dst)
````

````
kernel = np.float32([[2,-1,-1],
                    [-1,2,-1],
                    [-1,-1,2]])

dst= cv2.filter2D(image,-1,kernel)
cv2_imshow(dst)
````

##**Blur Kernel**

**Kernel:  **

```
            [1,2,1
             1,4,2    * 1/16
             1,2,1]
```

**Process :**

Its also called as a low pass filter and its objective is to reduce the intensity of edges .

````
kernel = np.float32([[1,2,1],
                     [2,4,2],
                     [1,2,1]])
kernel = (1/16)* kernel

dst= cv2.filter2D(image,-1,kernel)
cv2_imshow(dst)
````

#**Sharpen Kernel**

**Kernel:  **

```
[ 0,-1,0
 -1,5,-1
  0,-1,0]
```

**Process :**

Sharpens a image by increasing the edge content values also known as a high pass filter

````
kernel = np.float32([[0,-1,0],
                     [-1,5,-1],
                     [0,-1,0]])

dst= cv2.filter2D(image,-1,kernel)
cv2_imshow(dst)
````

#**Identity function**

**Kernel:  **

```
[0,0,0
 0,1,0
 0,0,0]
```

**Process :**

Output is same unchanged image as the anchor pixel is given as a output

````
kernel = np.float32([[0,0,0],
                     [0,1,0],
                     [0,0,0]])

dst= cv2.filter2D(image,-1,kernel)
cv2_imshow(dst)
````
